```python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib 

matplotlib.rc('xtick', labelsize=20) 
matplotlib.rc('ytick', labelsize=20) 
matplotlib.rcParams.update({'font.size': 20})
```

```python
workers=[8,24,32,64,128]
timing=[11.65,4.32,3.4,2.01,3.17]
timing_error=[0,0,0,0,0]
timing_scalable=[11.65,11.65/3,11.65/4,11.65/8,11.65/16]
```

```python
marker_style = dict(linestyle=':', color='cornflowerblue', markersize=10)

fig, ax = plt.subplots()
fig.figsize=(20, 8)
plt.errorbar(workers,timing,yerr=timing_error,fmt='o-',color='cornflowerblue',label='results')
plt.plot(workers,timing_scalable,'.-',marker='X',color='black',label='scalable')


plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xlabel('Number of workers')
plt.ylabel('Walltime in minutes')
plt.title('Performance tests of HAL - 192GB memory')
plt.savefig('figs/Perf-hal1.png', bbox_inches='tight')
```
