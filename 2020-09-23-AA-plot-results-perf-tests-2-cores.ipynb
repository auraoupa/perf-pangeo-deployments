```python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib 

matplotlib.rc('xtick', labelsize=10) 
matplotlib.rc('ytick', labelsize=10) 
matplotlib.rcParams.update({'font.size': 10})
```

```python
memory_cal=[8.39]
memory_gricad=[201]
memory_hal=[8,200]
memory_occ=[8,200]
cal=[221]
gricad=[71]
hal=[45,12]
occ=[58,49]
```

```python
marker_style = dict(linestyle=':', color='cornflowerblue', markersize=10)

fig, ax = plt.subplots()
fig.figsize=(20, 8)
plt.plot(memory_cal,cal,'.:',marker='X',color='cornflowerblue',label='CAL1')
plt.plot(memory_gricad,gricad,'.:',marker='X',color='seagreen',label='GRICAD')
plt.plot(memory_hal,hal,'.:',marker='X',color='purple',label='HAL')
plt.plot(memory_occ,occ,'.:',marker='X',color='darkorange',label='OCC')


plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xlabel('Memory in GB')
plt.ylabel('Walltime in minutes')
plt.title('Performance tests with 2 cores')
plt.savefig('figs/Perf-2cores.png', bbox_inches='tight')
```
