```python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib 

matplotlib.rc('xtick', labelsize=10) 
matplotlib.rc('ytick', labelsize=10) 
matplotlib.rcParams.update({'font.size': 10})
```

```python
workers=[2,8,24,32,64,128]
gricad=[59,20,13,10,10,10]
hal=[48,11,4,3,2,3]
occ=[49,15,5,4,2,1]
```

```python
marker_style = dict(linestyle=':', color='cornflowerblue', markersize=10)

fig, ax = plt.subplots()
fig.figsize=(20, 8)
plt.plot(workers,gricad,'.:',marker='X',color='seagreen',label='GRICAD')
plt.plot(workers,hal,'.:',marker='X',color='purple',label='HAL')
plt.plot(workers,occ,'.:',marker='X',color='darkorange',label='OCC')


plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xlabel('Number of Workers')
plt.ylabel('Walltime in seconds')
plt.title('Performance tests 200GB of memory')
plt.savefig('figs/Perf-200GB.png', bbox_inches='tight')
```
