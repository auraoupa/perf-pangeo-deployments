## 2 workers on every machine

```python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib 

matplotlib.rc('xtick', labelsize=20) 
matplotlib.rc('ytick', labelsize=20) 
matplotlib.rcParams.update({'font.size': 20})
```

```python
PC_workers2=[110]
PC_memory2=[33.66]
CAL_workers2=[199]
CAL_memory2=[8.39]
CLOUD_workers2=[77]
CLOUD_memory2=[8.59]
GRICAD_workers2=[107]
GRICAD_memory2=[201.35]
JZ_workers2=[57,42]
JZ_memory2=[128,340]
OCC_workers2=[51,57]
OCC_memory2=[240,120]
HAL_workers2=[45,48]
HAL_memory2=[60,30]
```

```python
marker_style = dict(linestyle=':', color='cornflowerblue', markersize=10)

fig, ax = plt.subplots()
fig.figsize=(20, 8)
plt.plot(PC_memory2,PC_workers2,'.',marker='X',label='PC',color='cornflowerblue')
plt.plot(CAL_memory2,CAL_workers2,'.',marker='X',label='CAL',color='seagreen')
plt.plot(CLOUD_memory2,CLOUD_workers2,'.',marker='X',label='CLOUD',color='purple')
plt.plot(GRICAD_memory2,GRICAD_workers2,'.',marker='X',label='GRICAD',color='darkorange')
plt.plot(JZ_memory2,JZ_workers2,'--',marker='X',label='JZ',color='palevioletred')
plt.plot(OCC_memory2,OCC_workers2,'--',marker='X',label='OCC',color='tomato')
plt.plot(HAL_memory2,HAL_workers2,'--',marker='X',label='HAL',color='gold')

plt.xlabel('Memory in GB')
plt.ylabel('Walltime in minutes')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Performance tests of PANGEO deployments - 2 workers asked')
plt.savefig('figs/Perf-2workers-all.png', bbox_inches='tight')
```

```python
marker_style = dict(linestyle=':', color='cornflowerblue', markersize=10)

fig, ax = plt.subplots()
fig.figsize=(20, 8)
plt.plot(JZ_memory2,JZ_workers2,'--',marker='X',label='JZ',color='palevioletred')
plt.plot(OCC_memory2,OCC_workers2,'--',marker='X',label='OCC',color='tomato')
plt.plot(HAL_memory2,HAL_workers2,'--',marker='X',label='HAL',color='gold')

plt.xlabel('Memory in GB')
plt.ylabel('Walltime in minutes')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Performance tests of HPC PANGEO deployments - 2 workers asked')
plt.savefig('figs/Perf-2workers-HPC.png', bbox_inches='tight')
```

## Results on the cloud

```python
CLOUD_timing=[77,39,18,10,4]
CLOUD_workers=[2,4,10,20,40]
CLOUD_memory=[8.59,17.18,42.95,85.9,171.8]
```

```python
marker_style = dict(linestyle=':', color='cornflowerblue', markersize=10)

fig, ax = plt.subplots()
fig.figsize=(25, 8)
ax.plot(CLOUD_workers,CLOUD_timing,'--',marker='X',label='CLOUD',color='purple')
ax2 = ax.twiny()
ax2.set_xlim(ax.get_xlim())
ax2.set_xticks(CLOUD_workers)
ax2.set_xticklabels(CLOUD_memory, rotation=45)
ax2.set_xlabel('Memory in GB', size=16)
ax2.tick_params(axis="x", bottom=False, top=True, labelbottom=False, labeltop=True, labelsize=10)
ax.set_xlabel('Number of workers', size=16)
ax.set_xticks(CLOUD_workers)
ax.set_ylabel('Walltime in minutes')
ax.tick_params(axis="x", bottom=True, top=False, labelbottom=True, labeltop=False)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Performance tests on PANGEO cloud', y=1.4)
plt.savefig('figs/Perf-cloud.png', bbox_inches='tight')
```
