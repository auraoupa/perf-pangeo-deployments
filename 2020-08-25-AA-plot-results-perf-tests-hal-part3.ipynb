```python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib 

matplotlib.rc('xtick', labelsize=20) 
matplotlib.rc('ytick', labelsize=20) 
matplotlib.rcParams.update({'font.size': 20})
```

```python
nodes=[5,10,20]
nodes_b2019=[5,10]
qdev1=[1203,664,355]
qdev1_sc=[1203,1203/2,1203/4]
qdev4=[304,179,116]
qdev4_sc=[304,304/2,304/4]
batch4=[300,192,146]
batch4_sc=[300,300/2,300/4]
qfull=[119,98,100]
qfull_sc=[119,119/2,119/4]
bfull=[113,93,117]
bfull_sc=[113,113/2,113/4]
b2019=[106,96]
b2019_sc=[106,106/2]
```

```python
marker_style = dict(linestyle=':', color='cornflowerblue', markersize=10)

fig, ax = plt.subplots()
fig.figsize=(10, 8)
plt.plot(nodes,qdev1,'.:',marker='X',color='cornflowerblue',label='qdev1 results')
plt.plot(nodes,qdev1_sc,'.-',marker='X',color='cornflowerblue',label='qdev1 scalable')
plt.plot(nodes,qdev4,'.:',marker='X',color='seagreen',label='qdev4 results')
plt.plot(nodes,qdev4_sc,'.-',marker='X',color='seagreen',label='qdev4 scalable')
plt.plot(nodes,batch4,'.:',marker='X',color='purple',label='batch4 results')
plt.plot(nodes,batch4_sc,'.-',marker='X',color='purple',label='batch4 scalable')
plt.plot(nodes,qfull,'.:',marker='X',color='darkorange',label='qfull results')
plt.plot(nodes,qfull_sc,'.-',marker='X',color='darkorange',label='qfull scalable')
plt.plot(nodes,bfull,'.:',marker='X',color='palevioletred',label='bfull results')
plt.plot(nodes,bfull_sc,'.-',marker='X',color='palevioletred',label='bfull scalable')
plt.plot(nodes_b2019,b2019,'.:',marker='X',color='tomato',label='b2019 results')
plt.plot(nodes_b2019,b2019_sc,'.-',marker='X',color='tomato',label='b2019 scalable')


plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xlabel('Number of nodes')
plt.ylabel('Walltime in seconds')
plt.title('Performance tests of HAL - increasing number of nodes')
plt.savefig('figs/Perf-hal3.png', bbox_inches='tight')
```
