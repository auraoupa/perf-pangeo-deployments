```python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib 

matplotlib.rc('xtick', labelsize=10) 
matplotlib.rc('ytick', labelsize=10) 
matplotlib.rcParams.update({'font.size': 10})
```

```python
nodes=[5,10,20]
hsw24=[75,52,38]
hsw24_sc=[75,75/2,75/4]
bdw28=[77,48,40]
bdw28_sc=[75,75/2,75/4]
```

```python
marker_style = dict(linestyle=':', color='cornflowerblue', markersize=10)

fig, ax = plt.subplots()
fig.figsize=(20, 8)
plt.plot(nodes,hsw24,'.:',marker='X',color='cornflowerblue',label='HSW24 results')
plt.plot(nodes,hsw24_sc,'.-',marker='X',color='cornflowerblue',label='HSW24 scalable')
plt.plot(nodes,bdw28,'.:',marker='X',color='seagreen',label='BDW28 results')
plt.plot(nodes,bdw28_sc,'.-',marker='X',color='seagreen',label='BDW28 scalable')


plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xlabel('Number of nodes')
plt.ylabel('Walltime in seconds')
plt.title('Performance tests of OCCIGEN - increasing number of nodes')
plt.savefig('figs/Perf-occ3.png', bbox_inches='tight')
```
